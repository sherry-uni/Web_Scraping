```python
# Dcard-Trending-香港
import requests, json
import pandas as pd
import re

article_id=[]
title_i_want=[]
content_for_each=[]
date=[]
article_all_id=[]
excerpt=[]

for j in range(1000):
    try: 
        last=str(article_all_id[-1])
    except:
        last=str(233914588)
    url="https://www.dcard.tw/_api/forums/trending/posts?popular=false&limit=100&before=" + last
    requ = requests.get(url)
    rejs=requ.json()
    for i in range(len(rejs)):
        article_all_id.append(rejs[i]['id'])
        if ('香港' in rejs[i]['title']) or ('香港' in rejs[i]['excerpt']):
            title_i_want.append(rejs[i]['title'])
            date.append(rejs[i]['createdAt'])
            article_id.append(rejs[i]['id'])
            excerpt.append(rejs[i]['excerpt'])
for id in article_id:
    url_content="https://www.dcard.tw/_api/posts/{id}/comments".format(id=id)
    requ = requests.get(url_content)
    rejs = requ.json()
    content=[]
    for i in range(len(rejs)):
        try:
            content.append(rejs[i]['content'])
        except:
                continue
    content_for_each.append(content)

dict={'article':article_id,'createdAt':date,'title':title_i_want,'excerpt':excerpt,'content':content_for_each}
df_trending=pd.DataFrame(dict)
df_trending
```

```python
df_trending.to_csv('dcard_trending_hk.csv')
```
